# My Submission Equation

```julia
f(x) = sin(pi*x) * cos(pi*x)
```

$f(x) = Sin({\pi}x).Cos({\pi}x)$\
A simple function that multiplies the trigonometric $Sine$ and $Cosine$ functions for a range of values in $x$

```julia
for x = -5:5
    println("[ x = ", x, " , f(x) = ", f(x), " ]")
end
```

```julia
using Plots
gr()
plot(f, -5, 5, legend = false, line = (:path, "blue"), title = "My equation plot", xlabel = "x", ylabel = "f(x)",
    marker = (:hex, 2, "red"), linewidth = 3, background = RGB(0.2, 0.2, 0.2))
```
